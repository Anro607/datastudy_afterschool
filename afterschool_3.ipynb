```python
"도서정보나루 인증키 : 5ac30542421fd1c33fc8de75c03a347f994093444c2039927322225298dace62"
```

```python
xml_a = """
<book>
      <name>고주연</name>
      <school>신성여자고등하교</school>
      <year>2030</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
x = et.fromstring(xml_a)
```

```python
x_child = list(x)
print(x_child)
```

```python
name = x.findtext('name')
school = x.findtext('school')
year = x.findtext('year')
print(name + school + year)
```

```python
import requests
a = '도리언 그레이의 초상'

url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_sug.pre&fbm=0&acr=1&acq=%EB%8F%84%EB%A6%AC%EC%96%B8%EA%B7%B8%EB%A0%88&qdt=0&ie=utf8&query={}'
r = requests.get(url.format(a))
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
prd_link = soup.find('a', attrs={'class':'more_multi'})
print(prd_link)
```

```python
print(prd_link['href'])
```

```python
url2 = 'https://terms.naver.com/search.naver?query='+prd_link['href']
r2 = requests.get(url2)
print(r2.text)
```
